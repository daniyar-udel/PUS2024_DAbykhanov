## Meausre the central tendency of the Philly park sizes


```python
meanps = np.mean(pparks['ACREAGE'])
medians = np.median(pparks['ACREAGE'])

print('Mean value of ACREAGE is:', meanps)
print('Median value of ACREAGE is:', medians)
```

### We can see the difference between Mean and Median so we can conclude that it is not normal distribuation because Mean and Median are not close to each one. We see this situation probably because we have outliers in our data

```python
fig = plt.figure() # Create a figure
ax = pparks["ACREAGE"].hist(bins = 30) # create a histogram for ACREAGE
ax.axvline(meanps, color = 'red', label = 'mean') # create a histogram for Mean
ax.axvline(medians, color = 'yellow', label = 'median') # create a histogram for Median
ax.set_ylabel('number of parks') # set y label
ax.set_xlabel('park size in acres') # set x label
ax.legend() # legend for graphic
plt.show() #show the graphic
```

### Here we can see spread of park size and the value of mean and median. We can say that it is poison distributed and we have a lot of outliers and extremly big values

## what is the spread of park sizes?


```python
stdps = np.std(pparks['ACREAGE']) # calculate stdv
print('Standart deviation of park size is:', stdps)
```

## select only a subset of the parks: the pasks smaller than 10 acres


```python
pparks_less_10 = pparks[pparks['ACREAGE'] < 10] # create df where park size less than 10
pparks_less_10.head() #first 5 rows
```

```python
meanps = np.mean(pparks_less_10['ACREAGE']) # calculate mean
medians = np.median(pparks_less_10['ACREAGE']) # calculate median

print('Mean value of ACREAGE is:', meanps)
print('Median value of ACREAGE is:', medians)
```

```python
fig = plt.figure() # Create a figure
ax = pparks_less_10["ACREAGE"].hist(bins = 30) # create a histogram for ACREAGE
ax.axvline(meanps, color = 'red', label = 'mean') # create a histogram for Mean
ax.axvline(medians, color = 'yellow', label = 'median') # create a histogram for Median
ax.set_ylabel('number of parks') # set y label
ax.set_xlabel('park size in acres') # set x label
ax.legend() # legend for graphic
plt.show() #show the graphic
```

###  This figure shows a spread of Park size with the value of mode and median.We can see that the Mean and Median become close to each other because we exclude outiers and extremly big values of park size



### Group by our data by Parent name and see the sum

```python
axx = pparks.groupby('PARENT_NAME')['ACREAGE'].sum() # groupby by parent_name and aggregate the sum
axx.head() # first 5 row
```

```python
print('Number of sample before grouping by PARENT NAME is:',pparks.shape[0]) # print the shape
print('Number of sample after grouping by PARENT NAME is:',axx.shape[0]) # print the shape
```

```python
meanps = np.mean(axx) # calculate mean
medians = np.median(axx) # calculate median
mode = axx.mode()[0] # calculate mode
stdv = np.std(axx) # calculate stdv

print('Mean value of ACREAGE is:', meanps)
print('Median value of ACREAGE is:', medians)
print('Mode value of ACREAGE is:', mode)
print('Stdv value of ACREAGE is:', stdv)
```

```python
fig = plt.figure() # Create a figure
ax = axx.hist(bins = 30) # create a histogram for ACREAGE
ax.axvline(meanps, color = 'red', label = 'mean') # create a histogram for Mean
ax.axvline(medians, color = 'yellow', label = 'median') # create a histogram for Median
ax.set_ylabel('number of parks') # set y label
ax.set_xlabel('park size in acres') # set x label
ax.legend() # legend for graphic
plt.show() #show the graphic
```

### We can see the main difference between there two datasets:

*  The mean value is more on number 6
*  The median value is less on 0.5
