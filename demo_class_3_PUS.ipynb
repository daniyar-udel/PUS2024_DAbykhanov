```python
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import norm
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
np.random.seed(312)
data = np.random.randn(200)
xr = np.linspace(-3, 3, 100)
plt.plot(xr, norm.pdf(xr, loc = 0, scale = 1))
plt.hist(data, density = 1)
plt.xlabel('some data')
plt.ylabel('frequency')
plt.plot()
```

```python
mean, std = stats.norm.fit(data)
```

```python
print(mean)
print(std)
```

```python
data.mean()
```

```python
data.std()
```

```python
data = np.random.poisson(1, 200)
xr = np.arange(0,5, 1)
plt.hist(data, density = 1)
plt.plot(xr, poisson.pmf(xr, mu = 1))
plt.show()
```
